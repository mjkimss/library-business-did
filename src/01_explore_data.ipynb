```python
def get_licenses_near(lat, lon, radius_meters, client):
    where_clause = (
        f"within_circle(location, {lat}, {lon}, {radius_meters}) "
        f"AND application_type = 'ISSUE'"
    )
    results = client.get(
        "r5kz-chrr",
        where=where_clause,
        limit=50000
    )
    return pd.DataFrame.from_records(results)
```

```python
test_row = treatment_df.iloc[0]  # Daley, Richard M.
test_licenses = get_licenses_near(test_row['lib_lat'], test_row['lib_long'], radius_meters=804, client=client)
# 804 meters ≈ 0.5 miles
print(len(test_licenses))
test_licenses.head()
```

```python
import time

all_results = []

for radius_label, radius_meters in [('0.25mi', 402), ('0.5mi', 804)]:
    for idx, row in treatment_df.iterrows():
        licenses = get_licenses_near(row['lib_lat'], row['lib_long'], radius_meters, client)
        licenses['branch_'] = row['branch_']
        licenses['radius_spec'] = radius_label
        all_results.append(licenses)
        print(f"{row['branch_']} ({radius_label}): {len(licenses)} records")
        time.sleep(0.5)  # be polite to the API

licenses_all = pd.concat(all_results, ignore_index=True)
print(f"\nTotal rows: {len(licenses_all)}")
```

```python
# Convert license_start_date to datetime and extract year-quarter
licenses_all['license_start_date'] = pd.to_datetime(licenses_all['license_start_date'])
licenses_all['year_quarter'] = licenses_all['license_start_date'].dt.to_period('Q')

# Aggregate: count new licenses per branch, per radius spec, per quarter
panel = (
    licenses_all
    .groupby(['branch_', 'radius_spec', 'year_quarter'])
    .size()
    .reset_index(name='new_business_count')
)

panel.head(10)
```

```python
# Build the full set of possible quarters across your data's range
all_quarters = pd.period_range(
    start=licenses_all['year_quarter'].min(),
    end=licenses_all['year_quarter'].max(),
    freq='Q'
)

# Build every combination of branch, radius spec, and quarter
from itertools import product
full_index = pd.DataFrame(
    list(product(treatment_df['branch_'], ['0.25mi', '0.5mi'], all_quarters)),
    columns=['branch_', 'radius_spec', 'year_quarter']
)

# Merge your actual counts onto this full grid, filling missing combos with 0
panel_full = full_index.merge(panel, on=['branch_', 'radius_spec', 'year_quarter'], how='left')
panel_full['new_business_count'] = panel_full['new_business_count'].fillna(0).astype(int)

panel_full.head(10)
```
